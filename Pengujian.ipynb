# Avalanche Effect

```python
def hamming_distance(str1, str2):
    str1 = str1.replace(" ", "")
    str2 = str2.replace(" ", "")
    
    hamming_dist = 0
    for i in range(min(len(str1), len(str2))):
        if str1[i] != str2[i]:
            hamming_dist += 1
    
    return hamming_dist

def avalanche_effect(str1, str2):
    hamming_dist = hamming_distance(str1, str2)
    max_len = max(len(str1.replace(" ", "")), len(str2.replace(" ", "")))
    avalanche_effect = (hamming_dist / max_len) * 100
    
    return avalanche_effect

def str_to_binary(string):
    binary_list = []
     
    for char in string:
        binary_list.append(bin(ord(char))[2:].zfill(8))
    return ''.join(binary_list)

def read_text_file(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    return text

file_pathoriginal = 'testing-ciphertext-original.txt'
file_path1 = 'testing-ciphertext-varian1.txt'
file_path2 = 'testing-ciphertext-varian2.txt'
file_path3 = 'testing-ciphertext-varian3.txt'

# Baca isi file ciphertext
ciphertextoriginal = read_text_file(file_pathoriginal)
ciphertext1 = read_text_file(file_path1)
ciphertext2 = read_text_file(file_path2)
ciphertext3 = read_text_file(file_path3)

# Ubah ke dalam bentuk biner
binary_ciphertextoriginal = str(str_to_binary(ciphertextoriginal))
binary_ciphertext1 = str(str_to_binary(ciphertext1))
binary_ciphertext2 = str(str_to_binary(ciphertext2))
binary_ciphertext3 = str(str_to_binary(ciphertext3))

avalanche_effect = avalanche_effect(binary_ciphertextoriginal, binary_ciphertext1)
print(f"Avalanche effect varian 1: {avalanche_effect:.2f}%")
```

# MSE dan PSNR

```python
import cv2
import numpy as np

def mse(original_img, stego_img):
    mse_value = np.mean((original_img - stego_img) ** 2)
    return mse_value

def psnr(original_img, stego_img):
    max_pixel = 255.0
    mse_value = mse(original_img, stego_img)
    if mse_value == 0:
        return float('inf')
    psnr_val = 10 * np.log10((max_pixel ** 2) / mse_value)
    return psnr_val

original_image = cv2.imread("pengujian/stego/cat/cover.png", cv2.IMREAD_GRAYSCALE)
stego_image = cv2.imread("pengujian/stego/cat/stego3.png", cv2.IMREAD_GRAYSCALE)
height, width = original_image.shape

stego_image = cv2.resize(stego_image, (width, height))

psnr_value = psnr(original_image, stego_image)
mse_value = mse(original_image, stego_image)

print("PSNR : ", psnr_value, " dB")
print ("MSE : ",mse_value)
```

# Recovery

```python
import hashlib

def calculate_hash(file_path, algo="sha256"):
    hash_object = hashlib.new(algo)
    with open(file_path, 'rb') as file:
        while True:
            data = file.read(4096)
            if not data:
                break
            hash_object.update(data)
    return hash_object.hexdigest()

def compare_hashes(file1_path, file2_path, algo="sha256"):
    hash_file1 = calculate_hash(file1_path, algo)
    hash_file2 = calculate_hash(file2_path, algo)

    if hash_file1 == hash_file2:
        return True
    else:
        return False

file1_path = "pengujian/stego/cat/testing-ciphertext-original.txt"
file2_path = "pengujian/stego/cat/testing-ciphertext-original3.txt"

# Cek apakah kedua file memiliki hash yang sama
if compare_hashes(file1_path, file2_path):
    print("Kedua file memiliki hash yang sama.")
else:
    print("Kedua file memiliki hash yang berbeda.")
```

```python
import cv2
import matplotlib.pyplot as plt

def plot_rgb_histogram(image_path):
    # Baca citra menggunakan OpenCV
    image = cv2.imread(image_path)

    # Ubah citra dari BGR menjadi RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Pisahkan saluran warna RGB
    red_channel = image_rgb[:,:,0]
    green_channel = image_rgb[:,:,1]
    blue_channel = image_rgb[:,:,2]

    # Hitung histogram untuk setiap saluran warna
    hist_red, bins_red = np.histogram(red_channel.ravel(), bins=256, range=[0, 256])
    hist_green, bins_green = np.histogram(green_channel.ravel(), bins=256, range=[0, 256])
    hist_blue, bins_blue = np.histogram(blue_channel.ravel(), bins=256, range=[0, 256])

    # Tampilkan grafik histogram
    plt.figure(figsize=(8, 5))
    plt.title("RGB Histogram")
    plt.xlabel("Intensitas Warna")
    plt.ylabel("Jumlah Piksel")
    plt.plot(hist_red, color='red', alpha=0.5)
    plt.plot(hist_green, color='green', alpha=0.5)
    plt.plot(hist_blue, color='blue', alpha=0.5)
    plt.legend(('Red', 'Green', 'Blue'))
    plt.show()

# Contoh pemanggilan fungsi
image_path = "pengujian/stego/pan-dog/stego3.png"  # Ganti dengan path citra Anda
plot_rgb_histogram(image_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img_path = 'pengujian/stego/pom/cover.png'
img = cv2.imread(img_path)

# Calculate image details using Laplacian filter
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
laplacian = cv2.Laplacian(gray_img, cv2.CV_64F)
avg_detail = np.mean(np.abs(laplacian))

# Calculate depth of color (number of unique colors)
unique_colors = len(np.unique(img))

# Calculate noise level
noise = np.mean(np.std(img, axis=0))

# Create a bar chart for the analysis
categories = ['Average Detail', 'Depth of Color', 'Noise Level']
values = [avg_detail, unique_colors, noise]

plt.bar(categories, values, color='red')
plt.title('Pom-pom')

plt.show()
```

```python
import numpy as np
from skimage import io, color
from skimage.measure import shannon_entropy

# Load image
image = io.imread('pengujian/stego/rick/cover.png')

# Convert image to grayscale
gray_image = color.rgb2gray(image)

# Convert image to binary representation
binary_image = np.unpackbits(np.uint8(gray_image * 255)).reshape(gray_image.shape + (-1,))

# Calculate bitplane complexity
bitplane_complexities = []
for bitplane in range(8):
    bitplane_data = binary_image[:, :, bitplane]
    bitplane_entropy = shannon_entropy(bitplane_data)
    bitplane_complexities.append(bitplane_entropy)

# Print the complexities of each bitplane
for bitplane, complexity in enumerate(bitplane_complexities):
    print(f"Bitplane {bitplane}: Complexity = {complexity}")
```
